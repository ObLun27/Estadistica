```python

import random
#Crear lista vacía que representa la edad de 100 personas
edades = []
#ciclo para generar 100 números aleatorios que representan la edad de las personas
for i in range(100):
  # generamos números aleatorios con randint entre 18 y 70
  numero_aleatorio = random.randint(18, 70)
  edades.append(numero_aleatorio)
print(edades)
```

```python
import pandas as pd
```

```python
#Crear un DataFrame de pandas con los números aleatorios
df = pd.DataFrame(edades, columns=['Edad'])
df.head()
```

```python
#Calcular la tabla de distribución de frecuencias
tabla_frecuencias = df['Edad'].value_counts().reset_index()
tabla_frecuencias.columns = ['Edad', 'Frecuencia']
tabla_frecuencias = tabla_frecuencias.sort_values(by='Edad')

#Imprimir la tabla
print("Tabla de distribución de frecuencias:")
tabla_frecuencias
```

```python
tabla_frecuencias['Frec Relativa'] = tabla_frecuencias['Frecuencia'] / 100
tabla_frecuencias
```

```python
tabla_frecuencias['Frec Porcentual'] = tabla_frecuencias['Frecuencia'] / 100 * 100
tabla_frecuencias
```

```python
tabla_frecuencias['Frec Acumulada'] = tabla_frecuencias['Frecuencia'].cumsum()
tabla_frecuencias
```

#Regla de Sturges
1. Es una regla que sirve para calcular el número o intervalos recomendado en el que se debe dividir un intervalos
c = 1 + log2(N)

```python

#importar la librería para funciones matemáticas
import math

print(edades)
```

```python
#crear un DataFrame con las edades
df = pd.DataFrame({'Edad': edades})
df
```

```python
#Calcular número de intervalos
n = len(edades)
K = int( 1 + math.log2(n))
print("Valor de n:", n , "\nNúmero de intervalos =", K)
```

```python
#Crear tabla de frecuencias
tabla_frecuencias = pd.cut( df['Edad'], bins=K, include_lowest=True).value_counts().reset_index()
tabla_frecuencias.columns = ['Clase', 'Frec']
tabla_frecuencias
```

```python
#Ordenar la tabla de frecuencias
tabla_frecuencias = tabla_frecuencias.sort_values(by='Clase')
tabla_frecuencias
```

```python
tabla_frecuencias['Frec Rel'] = tabla_frecuencias['Frec'] / n
tabla_frecuencias['Frec Porc'] = tabla_frecuencias['Frec'] / n * 100
tabla_frecuencias['Frec Ac'] = tabla_frecuencias['Frec'].cumsum()
tabla_frecuencias['Frec Rel Ac'] = tabla_frecuencias['Frec Rel'].cumsum()
tabla_frecuencias['Frec Porc Ac'] = tabla_frecuencias['Frec Porc'].cumsum()
#, Mostrar la tabla de frecuencia
print(tabla_frecuencias)
```

#Practica 5 Diagrama de Cajas y Bigotes

```python
import random
#Crear lista vacía que representa la edad de 100 personas
edades2 = []
#ciclo para generar 100 números aleatorios que representan la edad de las personas
for i in range(100):
  # generamos números aleatorios con randint entre 18 y 70
  numeroAleatorio = random.randint(20, 87)
  edades2.append(numeroAleatorio)
print(edades2)
```

```python
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(9,4))

bplot1 = ax1.boxplot(edades, vert = True, patch_artist=True)
ax1.set_title('Edades Encuesta 1')

boxplot2 = ax2.boxplot(edades2, vert = True, patch_artist=True)
ax2.set_title('Edades Encuesta 2')
plt.show()
```
