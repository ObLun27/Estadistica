```python
import math as mt # Para utilizar factorial
from itertools import permutations
from itertools import product
import numpy as np
```

```python
asistentes = ["a", "b", "c", "d", "e", "f", "g", "h", "i", "j"]
n = len(asistentes)
r = 4 #tamaño del subconjunto que representa las cuatro preguntas

pR = mt.factorial(n) / mt.factorial(n-r)
print("Existen ", pR, "posibles permutaciones sin repeticion del total de " ,
      n, "asistentes ", "en grupos de ", r ," en ", r)
print("Aqui las permutaciones sin repeticion son: ")
permutAsist = list(permutations(asistentes, r))
#Imprimir la lista de permutaciones
for permutacion in permutAsist:
    print(permutacion)
```
