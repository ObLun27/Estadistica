```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import pearsonr
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

##Grupo 1

```python
g1 = pd.read_csv('grupo1.csv')
g1.head()
```

```python
plt.figure(figsize=(6,4))
sns.histplot(g1["Calificacion"], bins=10, kde=True, color="skyblue")
plt.xlabel("Calificacion")
plt.ylabel("Horas de Estudio")
plt.title("Histograma de Calificacion del Grupo 1")
plt.show()
```

```python
#Regresion Lineal
# Regresin lineal
X1 = g1[["HorasEstudio"]].values
y1 = g1["Calificacion"].values
modelo1 = LinearRegression().fit(X1, y1)
pred1 = modelo1.predict(X1)

# Parmetros
pend1 = modelo1.coef_[0]
intercep1 = modelo1.intercept_
r1, _ = pearsonr(g1["HorasEstudio"], g1["Calificacion"])
r2_1 = r2_score(y1, pred1)

print(f"Ecuacin: y = {pend1:.2f}x + {intercep1:.2f}")
print(f"Coeficiente de correlacin de Pearson (r): {r1:.4f}")
print(f"Coeficiente de determinacin (R): {r2_1:.4f}")
```

```python
#Crear el diagrama de dispersion con linea de regresion
plt.figure(figsize=(6,4))
sns.scatterplot(x=g1["HorasEstudio"], y=g1["Calificacion"])
plt.plot(g1["HorasEstudio"], pred1, color="red", linewidth=2, label="Regresin Lineal")
plt.xlabel("Horas de Estudio")
plt.ylabel("Calificacion")
plt.grid(True)
plt.legend()
plt.title("Diagrama de Dispersin y Regresin Lineal del Grupo 1")
plt.show()
```

```python
#Coeficiente de correlacion
print("El coeficiente de correlacin (Pearson) para el grupo 1: r = ", f"{r1:.4f}")
```

**Cul es la interpretacin del coeficiente de regresin en trminos de la
relacin entre el tiempo de estudio y las calificaciones finales?**

Como se puede apreciar en el diagrama y en el valor del coeficiente de regresin, no se encuentra una relacin, ya que el valor del coeficiente es negativo lo que indica que las horas de estudio no son un determinante de la calificacion obtenida por los alumnos.

**Cmo se interpreta el coeficiente de correlacin en trminos de la fuerza y
direccin de la relacin entre las variables?**

No tiene relacin, obtuvo una correlacin negativa, adems, el coeficiente de determinacin indica un 33% de precisin del diagrama, por lo que las variables no estan estrechamente relacionadas

**El modelo de regresin lineal es estadsticamente significativo?**

No ya que como se mencion, es acertivo solo en un 33%, por lo que no sera de gran utilidad.

**Qu tan confiable es el modelo de regresin para predecir las
calificaciones finales de los estudiantes basado en el tiempo de estudio?**

 No, ya que no pose una correlacin fuerte y su coeficiente de determinacin es de tan solo 33%, por lo que no es factible utilizar este modelo para predecir calificaciones de los alumnos.
