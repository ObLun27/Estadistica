#Distribucion Normal
##Ejercicio 1

Una revision de los pesos de 500 estudiantes del ITD, evidenci que la media es de 70 kilogramos y la desviacion estandar es de 4 kilogramos. Suponiendo que los valores se distribuyen normalmente. Hallar lo siguiente:

* Generar aleatoriamente las edades de los 500 estudiantes con una media de 70kg y una desviacion estandar de 4kg
* Realizar un histograma de las edades.
* Graficar la funcin de densidad de probabilidad (PDF).

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Generar edades aleatorias con una media de 70kg y una desviacin estndar de 4kg
media = 70
desvEst = 4
nDatos = 500
edades = np.random.normal(media, desvEst, nDatos)
edades[0:50] # mostrar los primeros 50
```

```python
#Crear rango de valores para crear la densidad
rango = np.linspace(media - 3 * desvEst, media + 3 * desvEst, nDatos)
#Calcular la PDF de la distribucion normal
pdf = 1 / (desvEst * np.sqrt(2 * np.pi)) * np.exp(-(rango - media) ** 2 / (2 * desvEst ** 2))
print("Imprimir los primeros 50 valores del rango")
print(rango[0:50])
print("Imprimir los primeros 50 valores de la PDF")
print(pdf[0:50])
```

```python
plt.figure(figsize=(6, 4))
plt.hist(edades, bins=15, alpha=0.4, color='b')
plt.title('Distribucion Normal')
plt.xlabel("Peso en kg de los estudiantes")
plt.grid(True)
plt.show()
```

```python
#Visualizar la PDF
plt.figure(figsize=(6, 4))
plt.plot(rango, pdf, label='Normal:(media:{})'.format(np.round(media,2), np.round(desvEst,2)))
plt.title('Funcion de Densidad de Probabilidad (PDF) - Distribucion Normal')
plt.xlabel("Peso en kg de los estudiantes")
plt.ylabel("Densidad de Probabilidad")
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Graficar Histograma
plt.figure(figsize=(6, 4))
plt.hist(edades, bins=15, density=True, alpha=0.4, color='b')
plt.plot(rango, pdf, 'k', linewidth=2)
plt.title('Distribucion Normal')
plt.xlabel("Peso en kg de los estudiantes")
plt.grid(True)
plt.show()
```

#Ejercicio 2
Encontrar
* cuantos estudiantes pesan menos de 75kg.
* Cuantos estudiantes pesan menos de 62kg.
* Cuantos estuidantes pesan mas de 79kg.
Primero nececitamos calcular la probabilidad de que un estudiante pese menos de 75kg utilizando la funcion de distribucin acumulativa (CDF) de la distribucion normal.

```python
plt.figure (figsize=(6, 4))
media = 70
desvEst = 4
totalEstudiantes = 500
pesoLim = 79 #Variable para calcular la probabilidad de que un estudiante pese menos de 79kg

probMenos75 = norm.cdf(pesoLim, media, desvEst)
#Ahora calculamos el numero de estudiantes que pesan menos de 75kg
numMenos75 = probMenos75 * totalEstudiantes
print("Probabilidad de que un estudiante pese menos de 75kg", probMenos75)
print("Numero de estudiantes que pesan menos de 75kg:", numMenos75)

#Graficar la distribucion normal
x = np.linspace(media - 4 * desvEst, media + 4 * desvEst, 1000)
y = norm.pdf(x, media, desvEst)
plt.plot(x, y, 'b', linewidth=2)
plt.fill_between(x, y, where=(x <= pesoLim), color='skyblue', alpha=0.5)
plt.title('Probabilidad de que un estudiante pese menos de 75kg. \n $\mu=70$, $\sigma=4$, X = 500')
plt.xlabel("Peso en kg de los estudiantes")
plt.ylabel("Densidad de Probabilidad")
plt.grid(True)
plt.show()
```

##Cuantos estudiantes pesan mas de 79kg

```python
plt.figure (figsize=(6, 4))
media = 70
desvEst = 4
totalEstudiantes = 500
pesoLim = 79 #Variable para calcular la probabilidad de que un estudiante pese menos o igual de 79kg

probMenos79 = norm.cdf(pesoLim, media, desvEst)
probMas79 = 1 - probMenos79
#Ahora calculamos el numero de estudiantes que pesan menos de 75kg
numMas79 = probMas79 * totalEstudiantes
print("Probabilidad de que un estudiante pese mas de 79kg", probMas79)
print("Numero aproximado de estudiantes que pesan mas de 79kg:", numMas79)

#Graficar la distribucion normal
x = np.linspace(media - 4 * desvEst, media + 4 * desvEst, totalEstudiantes)
y = norm.pdf(x, media, desvEst)

plt.plot(x, y, 'b', linewidth=2)
plt.fill_between(x, y, where=(x >= pesoLim), color='skyblue', alpha=0.5)
plt.title('Probabilidad de que un estudiante pese menos de 75kg. \n $\mu=70$, $\sigma=4$, X = 500')
plt.xlabel("Peso en kg de los estudiantes")
plt.ylabel("Densidad de Probabilidad")
plt.grid(True)
plt.show()
```

#Practica 3. Problema de Distribucin binomial y de Poisson

Antes de resolver el problema especifique si la solucin es de tipo Binomial o Poisson y explique por qu.

Realizar un programa en Python que resuelva el siguiente problema:

El nmero x de personas ingresadas a una unidad de cuidados intensivos en un hospital particular, en un da, tiene una distribucin de probabilidad de Poisson con media igual a cinco personas por da.
a).Cul es la probabilidad de que el nmero de personas ingresadas a una unidad de cuidados intensivos en un hospital particular, en un da particular, sea dos?
b).Es probable que x exceda de 10? Explicar
##Tipo de distribucin:
Este problema es de tipo Poisson.

Por qu? Porque describe el nmero de eventos (personas ingresadas) que ocurren durante un intervalo en un tiempo fijo establecido (un da), y por que los eventos son independientes entre si, ocurren con una tasa de promedio ya conocida (es igual a 5) y se presentan de manera aleatoria.

```python
from scipy.stats import poisson
#Media (Lambda) de la distribucin de Poisson
media = 5

#Probabilidad de que x sea igual a 2 (osea que ingresen exactamente dos personas)
prob2Personas = poisson.pmf(2, media)

print("a) La probabilidad de que ingresen 2 personas:", prob2Personas)

#b) Es probable que x exceda de 10?
#La formula sera P(X > 10) = 1 - P(X <= 10)
probMas10 = 1 - poisson.cdf(10, media)

print("b) La probabilidad de que ingresen ms de 10 personas:", probMas10)

#Explicacion con un condicional para b)
if probMas10 < 0.05:
    print(" No es probable que ingresen ms de 10 personas (menos del 5%)")
else:
    print(" S es probable que ingresen ms de 10 personas (ms del 5%)")
```

Al calcular la probabilidad de que mas de 10 personas ingresen al hospital, se obtiene una probabilidad muy baja, por lo que no es posible que mas de 10 personas entren en un solo da.

##Problema 2
Antes de resolver el problema especifique si la solucin es de tipo Binomial o Poisson y explique por qu.

Un fabricantes de podadoras para el pasto compra motores de 1 hp y 2 ciclos, en lotes de 1,000, a un proveedor.
El fabricante equipa cada una de las podadoras producidas por su planta con uno de los motores.
La historia muestra que la probabilidad de que cualquier motor del proveedor resulte no satisfactorio es .001.
En un embarque de 1,000 motores, Calcular
a).Cul es la probabilidad de que ninguno sea defectuoso?
b).Cul es la probabilidad de que tres sean defectuosos?
c).Cul es la probabilidad de que cuatro sean defectuosos?

##Tipo de distribucion
Es de tipo Binomial ya que se tiene un nmero fijo de intentos (1,000)

Cada motor solo tiene dos posibles resultados (defectuoso o no defectuoso)

La probabilidad de que un motor sea defectuoso es constante (p = 0.001)

Los eventos son independientes

```python
from scipy.stats import binom

#Nmero de intentos (motores)
n = 1000

#Probabilidad de motor defectuoso
p = 0.001 #1%

#a) Probabilidad de que ningun motor sea defectuoso
prob0 = binom.pmf(0, n, p)

print("a) La probabilidad de que ninguno sea defectuoso:", prob0)

# b) probabilidad de que tres motores sean defectuosos
prob3 = binom.pmf(3, n, p)

print("b) La probabilidad de que tres motores sean defectuosos:", prob3)

# c) probabilidad de que cuatro motores sean defectuosos
prob4 = binom.pmf(4, n, p)

print("c) La probabilidad de que cuatro motores sean defectuosos:", prob4)
```
