```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import pearsonr
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

##Grupo 1

```python
g1 = pd.read_csv('grupo1.csv')
g1.head()
```

```python
plt.figure(figsize=(6,4))
sns.histplot(g1["Calificacion"], bins=10, kde=True, color="skyblue")
plt.xlabel("Calificacion")
plt.ylabel("Horas de Estudio")
plt.title("Histograma de Calificacion del Grupo 1")
plt.show()
```

```python
#Regresion Lineal
# Regresión lineal
X1 = g1[["HorasEstudio"]].values
y1 = g1["Calificacion"].values
modelo1 = LinearRegression().fit(X1, y1)
pred1 = modelo1.predict(X1)

# Parámetros
pend1 = modelo1.coef_[0]
intercep1 = modelo1.intercept_
r1, _ = pearsonr(g1["HorasEstudio"], g1["Calificacion"])
r2_1 = r2_score(y1, pred1)

print(f"Ecuación: y = {pend1:.2f}x + {intercep1:.2f}")
print(f"Coeficiente de correlación de Pearson (r): {r1:.4f}")
print(f"Coeficiente de determinación (R²): {r2_1:.4f}")
```

```python
#Crear el diagrama de dispersion con linea de regresion
plt.figure(figsize=(6,4))
sns.scatterplot(x=g1["HorasEstudio"], y=g1["Calificacion"])
plt.plot(g1["HorasEstudio"], pred1, color="red", linewidth=2, label="Regresión Lineal")
plt.xlabel("Horas de Estudio")
plt.ylabel("Calificacion")
plt.grid(True)
plt.legend()
plt.title("Diagrama de Dispersión y Regresión Lineal del Grupo 1")
plt.show()
```

```python
#Coeficiente de correlacion
print("El coeficiente de correlación (Pearson) para el grupo 1: r = ", f"{r1:.4f}")
```

**¿Cuál es la interpretación del coeficiente de regresión en términos de la
relación entre el tiempo de estudio y las calificaciones finales?**

Como se puede apreciar en el diagrama y en el valor del coeficiente de regresión, no se encuentra una relación, ya que el valor del coeficiente es negativo lo que indica que las horas de estudio no son un determinante de la calificacion obtenida por los alumnos.

**¿Cómo se interpreta el coeficiente de correlación en términos de la fuerza y
dirección de la relación entre las variables?**

No tiene relación, obtuvo una correlación negativa, además, el coeficiente de determinación indica un 33% de precisión del diagrama, por lo que las variables no estan estrechamente relacionadas

**¿El modelo de regresión lineal es estadísticamente significativo?**

No ya que como se mencionó, es acertivo solo en un 33%, por lo que no sería de gran utilidad.

**¿Qué tan confiable es el modelo de regresión para predecir las
calificaciones finales de los estudiantes basado en el tiempo de estudio?**

 No, ya que no pose una correlación fuerte y su coeficiente de determinación es de tan solo 33%, por lo que no es factible utilizar este modelo para predecir calificaciones de los alumnos.

```python
#Coeficiente de correlacion
print("El coeficiente de correlación (Pearson) para el grupo 1: r = ", f"{r2:.4f}")
```

**¿Cuál es la interpretación del coeficiente de regresión en términos de la
relación entre el tiempo de estudio y las calificaciones finales?**

En este caso, se logra apreciar una mayor relación entre el tiempo de estudio y las calificaciones finales de los alumnos, teniendo un coeficiente de correlación de 0.86, por lo que las horas de estudio si se ven reflejadas en las calificaciones en la mayoria de casos de este grupo.

**¿Cómo se interpreta el coeficiente de correlación en términos de la fuerza y
dirección de la relación entre las variables?**

La correlación obtenida fue del 0.86, por lo que se puede interpretar que los estudiantes que más estudiaron, obtuvieron tambien las calificaciones mas altas, por lo que hay una correlación más marcada que en el grupo anterior.

**¿El modelo de regresión lineal es estadísticamente significativo?**

Si, se obtuvo un coeficiente de determinación de 0.74, que puede interpretarse que este modelo es acertivo en un 74% de los casos.

**¿Qué tan confiable es el modelo de regresión para predecir las
calificaciones finales de los estudiantes basado en el tiempo de estudio?**

A pesar de no ser totalmente confiable, si puede acercarse más a una predicción más acertada de las calificaciones finales de los alumnos basandonos en sus horas dedicadas al estudio.
